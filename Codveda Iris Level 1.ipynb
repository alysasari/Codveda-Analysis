```python
df = pd.read_csv('iris.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
# dataset clear
```

```python
df.describe()
```

```python
# check distribution target class
print(df['species'].value_counts())
sns.countplot(x='species', data=df)
plt.show()
```

```python
#correlation fitur
sns.pairplot(df, hue='species')
plt.show()
```

```python
#barplot for species
plt.figure(figsize=(6,4))
sns.countplot(x='species', data=df, palette='viridis')
plt.xlabel('Species')
plt.ylabel('Total')
plt.title('Distribution Species')
plt.savefig('species.png')
plt.show()
#scatter plot for sepal length and sepal width
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, palette='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.title('Relation Sepal Length & Sepal Width')
plt.savefig('sepal.png')
plt.show()
#line plot
plt.figure(figsize=(6,4))
for species in df['species'].unique():
  data = df[df['species'] == species]
  plt.plot(data.index, data['sepal_length'], label = species )
plt.title('Correlation between sepal length in species')
plt.xlabel('index')
plt.ylabel('Sepal Length')
plt.legend()
plt.savefig('line.png')
plt.show()
```
